```python
def read_csv(choice):
    print("Please follow required csv input layout")
    if choice == "Make a list of wells and sample names into a plate visual layout":
        im = Image.open("pngs/96_input_list_template.png")
        im.show()
    elif choice == "Take a plate visual and make into a sample list with well location":
        im = Image.open("pngs/Plate_input_template.png")
        im.show()
    elif choice == "Rearray samples in A1, A2, A3 format to A1, B1, C1 format":
        im = Image.open("pngs/A1_A2_A3_template.png")
        im.show()
    elif choice == "Rearray samples in A1, B1, C1 format to A1, A2, A3 format":
        im = Image.open("pngs/A1_B1_C1_template.png")
        im.show()
    elif choice == "Replace A01 with A1 and so on":
        im = Image.open("pngs/A01_B01_Template.png")
        im.show()
    elif choice == "Rearray up to four 96 well plates into 384 plate":
        im = Image.open("pngs/384.png")
        im.show()
        return "Other_script"
    
    file_name = input("Type your csv name (e.g. '96_well_list_A1_A2.csv'):")
    try:
        csv_file = open(file_name, 'r')
        for line in record_file:
            print(line)
    except Exception as e:
        if "utf-8" in str(e):
            print("You haven't chosen a csv type file. File format must be a csv.")
            read_csv()
        elif "Errno 2" in str(e):
            try:
                csv_file = open(file_name + ".csv", 'r')
                for line in record_file:
                    print(line)
            except Exception as e:
                print(e)
                read_csv()

        return csv_file
```

```python
def drop_down_gui():
    
    OPTIONS = [
        "Make a list of wells and sample names into a plate visual layout",
        "Take a plate visual and make into a sample list with well location",
        "Rearray samples in A1, A2, A3 format to A1, B1, C1 format",
        "Rearray samples in A1, B1, C1 format to A1, A2, A3 format",
        "Replace A01 with A1 and so on",
        "Rearray up to four 96 well plates into 384 plate",
    ] 

    Sequencer = Tkinter.Tk()

    variable = Tkinter.StringVar(Sequencer)
    variable.set(OPTIONS[0]) # default value

    m = Tkinter.Label(Sequencer, text="How can I help rearray?")
    w = Tkinter.OptionMenu(Sequencer, variable, *OPTIONS)
    m.pack()
    w.pack()
    
    picked = []
    Close = Tkinter.Label(Sequencer, text="Great! Now please click the X to close.")
    
    def ok():
        stuff = variable.get()
        picked.append(str(stuff))
        Close.pack()
        
    button = Tkinter.Button(Sequencer, text="Submit", command=ok)
    button.pack()
    
    Sequencer.mainloop()
    return picked[0]
```

```python
def rearray_A1_B1(csv_file):
    dictionary_samples = {}
    
    for line in csv_file:
        line = line.rstrip("\n").split(",")
        dictionary_samples[line[0]] = line[1]
    
    
    
    As = {}
    Bs = {}
    Cs = {}   
    Ds = {}
    Es = {}
    Fs = {}
    Gs = {}
    Hs = {}    
    
    
    for key, value in dictionary_samples.items():
        
        if "A" in key:
            As[key] = value

        if "B" in key:
            Bs[key] = value
            
        if "C" in key:
            Cs[key] = value
             
        if "D" in key:
            Ds[key] = value
    
    
        if "E" in key:
            Es[key] = value
               
            
        if "F" in key:
            Fs[key] = value
            
        if "G" in key:
            Gs[key] = value
            
        if "H" in key:
            Hs[key] = value
            
        
    filename = "A1_A2_A3.csv"
    with open(filename, 'w') as filename:
        wr = csv.writer(filename, quoting=csv.QUOTE_ALL)
        for row in As.items():
            wr.writerow(row)
        for row in Bs.items():
            wr.writerow(row)
        for row in Cs.items():
            wr.writerow(row)
        for row in Ds.items():
            wr.writerow(row)
        for row in Es.items():
            wr.writerow(row)
        for row in Fs.items():
            wr.writerow(row)
        for row in Gs.items():
            wr.writerow(row)
        for row in Hs.items():
            wr.writerow(row)
```

```python
def replace_A01_with_A1(csv_file):
    dictionary_samples = {}
    
    list_of_good = ["A10", "B10", "C10", "D10", "E10", "F10", "G10", "H10"]
    
    for line in csv_file:
        line = line.rstrip("\n").split(",")
        if line[0] not in list_of_good:
            for char in line[0]:
                if char == "0":
                    line[0] = line[0].replace('0', "")
                    
        dictionary_samples[line[0]] = line[1]
  

    filename = "A01_changed_to_A1.csv"
    with open(filename, 'w') as filename:
        wr = csv.writer(filename, quoting=csv.QUOTE_ALL)
        for row in dictionary_samples.items():
            wr.writerow(row)
```

```python
def _96_to_384(quadrant1, quadrant2, quadrant3, quadrant4):
    
    #This makes up lists of each new well assignment in the 384 well plate, based on quadrants
    map1 =  []
    map1file = open("Required_to_Work/Quadrantmap1.csv", 'r')
    for line in map1file:
        line = line.replace("\n", "")
        map1.append(line) 
    map2 =  []
    map2file = open("Required_to_Work/Quadrantmap2.csv", 'r')
    for line in map2file:
        line = line.replace("\n", "")
        map2.append(line)
    map3 =  []
    map3file = open("Required_to_Work/Quadrantmap3.csv", 'r')
    for line in map3file:
        line = line.replace("\n", "")
        map3.append(line)
    map4 =  []
    map4file = open("Required_to_Work/Quadrantmap4.csv", 'r')
    for line in map4file:
        line = line.replace("\n", "")
        map4.append(line)    
    
    #turn into a dictionary, then if A1, replace it with whatever quadrant assignment will be, then combine
    #them all, then sort them
    
    #Q1
    quadrant1dictionary = {}
    for item in quadrant1:
        item = item.replace("\n", "")
        item = item.split(",")
        quadrant1dictionary[item[0]] = item[1]
    
    quadrant1dictionary.pop("\ufeffWell")
    
    #Q2
    quadrant2dictionary = {}
    for item in quadrant2:
        item = item.replace("\n", "")
        item = item.split(",")
        quadrant2dictionary[item[0]] = item[1]
    
    quadrant2dictionary.pop("\ufeffWell")
    
    #Q3
    quadrant3dictionary = {}
    for item in quadrant3:
        item = item.replace("\n", "")
        item = item.split(",")
        quadrant3dictionary[item[0]] = item[1]
    
    quadrant3dictionary.pop("\ufeffWell")
    
    
    #Q4
    quadrant4dictionary = {}
    for item in quadrant4:
        item = item.replace("\n", "")
        item = item.split(",")
        quadrant4dictionary[item[0]] = item[1]
    
    quadrant4dictionary.pop("\ufeffWell")
    
    #Q1    
    final_quad_1_dictionary = {}
    count = 0
    for key, value in quadrant1dictionary.items():
        final_quad_1_dictionary[map1[count]] = value
        count = count +1
    
    #Q2    
    final_quad_2_dictionary = {}
    count = 0
    for key, value in quadrant2dictionary.items():
        final_quad_2_dictionary[map2[count]] = value
        count = count +1
        
    #Q3    
    final_quad_3_dictionary = {}
    count = 0
    for key, value in quadrant3dictionary.items():
        final_quad_3_dictionary[map3[count]] = value
        count = count +1
    
    #Q4  
    final_quad_4_dictionary = {}
    count = 0
    for key, value in quadrant4dictionary.items():
        final_quad_4_dictionary[map4[count]] = value
        count = count +1
                            
    #Write all quadrants to the same csv and output
    
    filename = "384_rearray.csv"
    with open(filename, 'w') as filename:
        wr = csv.writer(filename, quoting=csv.QUOTE_ALL)
        for row in final_quad_1_dictionary.items():
            wr.writerow(row)
        for row in final_quad_2_dictionary.items():
            wr.writerow(row)
        for row in final_quad_3_dictionary.items():
            wr.writerow(row)
        for row in final_quad_4_dictionary.items():
            wr.writerow(row)    
    
   # new = pd.DataFrame.from_dict(dictionary_A, orient ='index')
    

    header_list = ["Well", "Sample_Name"]
    df = pd.read_csv("384_rearray.csv", names=header_list)
  
    df["Well"] = df["Well"].str.replace("\ufeff","")
    df = df.reindex(index=order_by_index(df.index, index_natsorted(df.Well)))
 
    df.to_csv("384_rearray.csv", index=False)
```
